```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
url = "https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv" 
df = pd.read_csv(url)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X= df.drop(columns=['target'])
y= df['target']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
models=[SVC(),RandomForestClassifier(random_state=42),LogisticRegression(max_iter=1000)]

for m in models:
    m.fit(X_train,y_train)
    predicttion=m.predict(X_test)
    print(f'testing_effiency',m.score(X_test,y_test))
    print(f'training_effiency',m.score(X_train,y_train))
    
```

```python
model=LogisticRegression(max_iter=2000)
model.fit(X_train,y_train)
pred=m.predict(X_train)
pred
print('testing_score',model.score(X_test,y_test))
print('training_score',model.score(X_train,y_train))
```
